```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
```

```python
N = 100000
A = np.linalg.inv(np.array([[5, 4],
                            [4, 5]]))
```

```python
samples = []
x = 0
y = 0
for _ in range(N):
    x = rng.normal()
    x = np.sqrt(1 / A[0, 0]) * x - A[0, 1] * y / A[0, 0]
    y = rng.normal()
    y = np.sqrt(1 / A[1, 1]) * y - A[0, 1] * x / A[1, 1]
    samples.append([x, y])
samples = np.array(samples)
plt.scatter(samples[:, 0], samples[:, 1])
```
